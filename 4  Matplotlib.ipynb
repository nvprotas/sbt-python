## 1
%matplotlib inline позволяет отрисовывать графики под ячейкой с кодом

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

## 2 и 3

```python
X = np.linspace(-5,5,100)

y = [np.abs(x) for x in X]
plt.plot(X,y, label = "$abs(x)$")

y = [np.sin(x) for x in X]
plt.plot(X,y, label = "$sin(x)$")

y = [-np.arctan(x) for x in X]
plt.plot(X,y, label = "$-arctan(x)$")

plt.title("$Good title$")
plt.xlabel("$X$")
plt.ylabel("$Y$")

plt.savefig("my beautiful fig")

plt.legend()
```

## 4

```python
plt.rcParams['figure.figsize'] = 10, 7
plt.subplot(221)
y = [x**2 for x in X]
plt.title("$x^2$")
plt.plot(X,y)

plt.subplot(222)
y = [x**3 for x in X]
plt.title("$x^3$")
plt.plot(X,y)

plt.subplot(223)
y = [np.exp(x) for x in X]
plt.title("$e^x$")
plt.plot(X,y)

plt.subplot(224)
y = [np.log(x) for x in X]
plt.title("$log(x)$")
plt.plot(X,y)
```
