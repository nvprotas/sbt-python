# 5 Всё вместе + SciPy

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return x**3+2*x**2 -3*x + 2

np.random.seed(42)
points_x = np.linspace(-10, 10, 500)
points_y = np.array([f(x) + np.random.normal(0, 100) for x in points_x])
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(points_x, points_y)
plt.grid()
plt.show()
```

```python
coef = np.polyfit(points_x,points_y,3)
points_y_fit = np.array([coef[0]*x**3+coef[1]*x**2 + coef[2]*x + [3] for x in points_x])
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(points_x, points_y)
plt.plot(points_x,points_y_fit, linewidth = 5, c = 'r')
plt.grid()
plt.show()
```

```python
def derivative(x):
    return 3*coef[0]*x**2+2*coef[1]*x**1 + coef[2]

def derivative_opt(x):
    return derivative(x)**2
```

```python
extr1 = optimize.minimize(derivative_opt, 2,method='nelder-mead').x[0]
y_extr1 = f(extr1)
extr2 = optimize.minimize(derivative_opt, -2,method='nelder-mead').x[0]
y_extr2 = f(extr2)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(points_x, points_y,alpha = 0.5)
plt.plot(points_x,points_y_fit, linewidth = 7, c = 'r',alpha = 0.5)
plt.scatter(extr1,y_extr1,c = 'r', linewidth = 7)
plt.scatter(extr2,y_extr2,c = 'r', linewidth = 7)
plt.grid()
plt.show()
```
